* We can now perform some text analysis on this dataset using either Python or R. With IPython Notebook, we can execute R commands in the shell as shown below. 


* The R code generates wordclouds for each of the sessions. This involves R's ``tm`` package to pre-process the text corpus that we have just created and the ``worcloud`` package to draw the wordcloud visualizations.

```python
# pip install rpy2
%pylab inline
%load_ext rmagic
```

```python
%%R 
 library(tm);
 library(wordcloud);
# for(z in 66:69){ 
for(z in 68:68){ 
     print(z)
     dirName = sprintf("UNGA_%d",z)
     print(dirName)     
     speeches<-Corpus(DirSource(dirName,pattern="*.pdf.txt"));
     speeches<-tm_map(speeches, stripWhitespace); 
     speeches<-tm_map(speeches,tolower);
     speeches<-tm_map(speeches,removePunctuation);
     speeches<-tm_map(speeches,removeNumbers);
         
     # Some observations from the 66-th session: palestine, israel, sudan, africa, libya
     # 67-th session: syria appears, women appears, has arab
     # 68-th session: syria becomes bigger, chemical appears, mdgs appears
     #speeches<-tm_map(speeches,stemDocument,language="english");
     ##R inspect(speeches)
     speeches<-tm_map(speeches,removeWords,stopwords("english"))
     print(sprintf("Wordcloud for speeches from the General Debate of the %d-th U.N. General Assembly session",z))
     #%R dev.print
     #png(sprintf("UNGA%d_wordcloud.png",z), width=1280,height=800,res=180)
     png(sprintf("UNGA%d_wordcloud.png",z), width=1280,height=800,res=180)
     wordcloud(speeches, scale=c(2,0.5), max.words=120, random.order=FALSE, rot.per=0.35, colors=brewer.pal(7,"Dark2"))
     dev.off()
     
     # with stopwords
     speeches<-tm_map(speeches,removeWords,c("president","international","nations","development",
                                             "security","united","countries","humans","human","people",
                                             "states","ago","ladies","gentlemen","must","will","well",
                                             "also","general","take","new","york","world","peace",
                                             "global","charter","agenda","conference","september",
                                             "minister", "thank", "great", "years","let","order","full",
                                             "address","last","including",
                                             stopwords("english"))); 
     #print(sprintf("Wordcloud for speeches from the General Debate of the %d-th U.N. General Assembly session",z))
     png(sprintf("UNGA%d_wordcloud_sq.png",z), width=550,height=500,res=180)
     wordcloud(speeches, scale=c(1,0.5), max.words=120, random.order=FALSE, rot.per=0.35, colors=brewer.pal(7,"Dark2"))
     dev.off()
 }
```

### Ex: Scrape U.S. Congressional Record PDFs

* Now let's try using ``PDFMiner`` on the U.S. Congressional Record PDFs which we downloaded earlier. Remember that these PDF files are stored in the ``archive/`` directory. The code to do this is listed below. 


* Inspect the text files that were created. How would you rate the quality of PDF scraping in this case?

```python
import commands
import glob

congressionalPDFs = glob.glob("archive/*.pdf")
print(len(congressionalPDFs))
for pdfFile in congressionalPDFs:
    print "Processing file " + pdfFile
    txtFile = pdfFile + ".txt"  # the output text file will have .pdf.txt extension
    status, output = commands.getstatusoutput("pdf2txt.py " + pdfFile)
    with open(txtFile,"w") as fout:
        fout.write(output)
```

* In general, ``pdf2txt.py`` doesn't work well on multi-column PDF documents since all the words in the different columns will end up getting mixed together. However, if the subsequent analysis assumes a bag of words model, then this does not pose problems. 


* Tables in PDF documents are another difficult feature to extract text from or perform optical character recognition (OCR). 


* Finally, the OCR process is inherently very sensitive to the quality of the PDF file. Performing high quality OCR requires a lot of skill and experience - this is one instance where proprietary commercial OCR software might be advisable. Some powerful open-source tools are also available, like tesseract-ocr (https://code.google.com/p/tesseract-ocr/) which is a software project supported by Google.
